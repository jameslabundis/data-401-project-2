```python
volumeTarget = df
yVal3 = volumeTarget["Volume_Per_Capita"]
volumeTarget = volumeTarget.drop("Volume_Per_Capita", axis = 1)
volumeTarget = volumeTarget.drop("Sales_Per_Capita", axis = 1)
volumeTarget = volumeTarget.drop("Bottles_Per_Capita", axis = 1)
```

```python
def SSE(X, y, betaHat):
    return np.sum((y-np.matmul(X, betaHat))**2)
```

```python
print(SSE(salesTarget, yVal1, reg1.coef_)/salesTarget.shape[0])
```

```python
print(SSE(bottlesTarget, yVal2, reg2.coef_)/bottlesTarget.shape[0])
```

```python
print(SSE(volumeTarget, yVal3, reg3.coef_)/volumeTarget.shape[0])
```

```python
def find_zero_coef(startList, coef): 
    for i in range(len(coef)):
        if coef[i] == 0:
            startList.append(i)
    
```

```python
dropListSales = []
find_zero_coef(dropListSales, reg1.coef_)
```

```python
salesTarget = salesTarget.drop(salesTarget.columns[dropListSales], axis = 1)
```

```python
salesFinalCoef = np.delete(reg1.coef_, dropListSales)
```

```python
#k = number of variables
#n = number of observations
#BIC = n*ln(sse/n) + k*ln(n) = number of variables
salesBIC = salesTarget.shape[0]*np.log(SSE(salesTarget, yVal1, salesFinalCoef)/salesTarget.shape[0]) + salesTarget.shape[1]*np.log(salesTarget.shape[0])
print(salesBIC)
```

```python
dropListBottles = []
find_zero_coef(dropListBottles, reg2.coef_)
```

```python
bottlesTarget = bottlesTarget.drop(bottlesTarget.columns[dropListBottles], axis = 1)
```

```python
bottlesFinalCoef = np.delete(reg2.coef_, dropListBottles)
```

```python
bottleBIC = bottlesTarget.shape[0] * np.log(SSE(bottlesTarget, yVal2, bottlesFinalCoef)/bottlesTarget.shape[0]) + bottlesTarget.shape[1]*np.log(bottlesTarget.shape[0])
print(bottleBIC)
```
